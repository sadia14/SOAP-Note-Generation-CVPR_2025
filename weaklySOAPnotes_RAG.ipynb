```python
import pandas as pd
import os
from docx import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
import ollama

# Load original image-caption data
df = pd.read_csv("fine_tune_data.csv")
image_folder = "/content/all_images"
df["image_path"] = df["img_id"].apply(lambda x: os.path.join(image_folder, x))

# Create soap_note column if missing
if "soap_note" not in df.columns:
    df["soap_note"] = ""

# Load and process the clinical document for retrieval
document_path = '/content/RAG_new.docx'
doc = Document(document_path)
document_text = "\n".join([p.text for p in doc.paragraphs])

# Split document into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
documents = text_splitter.create_documents([document_text])

# Create embedding-based retriever using OpenAI
embeddings = OpenAIEmbeddings(openai_api_key="Replace with your OpenAI API key")
vector_store = Chroma.from_documents(documents, embeddings, persist_directory="./chroma_store")
retriever = vector_store.as_retriever(search_type="similarity", search_kwargs={"k": 3})

# Prompt template for SOAP note generation
prompt_template = """
You are a clinical assistant AI. Given a dermoscopic lesion image, a caption describing lesion features, and relevant medical context, generate a structured and comprehensive SOAP (Subjective, Objective, Assessment, Plan) note.

---

**Subjective (S):**
- **Chief Complaint:** State the patient’s primary concern or reason for consultation.
- **Medical History:** Summarize relevant medical history (e.g., skin lesions, sun exposure, family history).

**Objective (O):**
- **On Examination (Image 1):** Include lesion shape, border, symmetry, pigmentation, vascular visibility.
- **Observation:** Note lesion size, surface texture, color variety, and symmetry.

**Assessment (A):**
- **Investigation & Diagnosis:** Provide likely diagnosis and differentials.
- **Summary:** Mention ABCDE criteria if applicable.

**Plan (P):**
- **Clinical Action:** Suggest next steps (e.g., biopsy, dermoscopy, referral).
- **Patient Advice:** Offer sun protection tips, monitoring guidance, and follow-up timing.

---

**Caption (Image Summary):** "{caption}"
**Context:** {context}
"""

# Function to generate SOAP note
def generate_clinical_note(caption, image_path):
    query = f"""
    Retrieve the information to generate a SOAP clinical note for a lesion image based on the following caption: {caption}.
    """
    context_docs = retriever.get_relevant_documents(query)
    context = "\n".join([doc.page_content for doc in context_docs])
    formatted_prompt = prompt_template.format(caption=caption, context=context)

    response = ollama.chat(
        model='llama3.2-vision',
        messages=[{
            'role': 'user',
            'content': formatted_prompt,
            'images': [image_path]
        }]
    )

    if 'message' in response and 'content' in response['message']:
        return response['message']['content']
    else:
        return "Error: No response from the model."

# Output CSV to save progress
output_csv = "fine_tune_data_with_soap_notes.csv"

# If progress exists, resume from there
if os.path.exists(output_csv):
    df_existing = pd.read_csv(output_csv)
    df = df.set_index("img_id").combine_first(df_existing.set_index("img_id")).reset_index()

# Loop through rows, generate & print SOAP note, and save
for idx, row in df.iterrows():
    if isinstance(row["soap_note"], str) and len(row["soap_note"].strip()) > 10:
        print(f"⏩ Skipping: {row['img_id']} (already processed)")
        continue

    print(f"\n🩺 Generating SOAP note for image: {row['img_id']}")
    try:
        caption = row["generated_caption"]
        image_path = row["image_path"]
        note = generate_clinical_note(caption, image_path)

        print("\n📝 Generated SOAP Note:\n")
        print(note)

        df.at[idx, "soap_note"] = note
        df.to_csv(output_csv, index=False)
        print(f"✅ Saved SOAP note for: {row['img_id']}")
    except Exception as e:
        print(f"❌ Failed for {row['img_id']}: {e}")

# Optional: Download the CSV
from google.colab import files
files.download(output_csv)
```
